```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_set = pd.read_csv("/content/Dataset_master - Salary Data Random Forest Regre.csv")
```

```python
data_set.describe()
```

```python
x=data_set.iloc[:,1:-1].values
y=data_set.iloc[:,-1].values
```

```python
print(x)
print(y)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=12,random_state=21)
regressor.fit(x,y)
```

```python
regressor.predict([[5.5]])
```

```python
x_grid = np.arange(min(x),max(x),0.01)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color='blue')
plt.title('random forest regression')
plt.xlabel('Grade')
plt.ylabel('salary')
plt.show()
```
