#Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#Dataset reading

```python
data_set = pd.read_csv("/content/Dataset_master - Salary Data Support Vector regr.csv")
```

```python
data_set.describe()
```

```python
x=data_set.iloc[:,1:-1].values
y=data_set.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

#feature Scaling

```python
y=y.reshape(len(y),1)
print(y)
```

#Training SVR Model on whole dataset

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()

X = sc_x.fit_transform(x)
Y = sc_y.fit_transform(y)
#print(X)
print(Y)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X,Y)
```

```python
regressor.predict(sc_x.transform([[6.5]]))
```

```python
plt.plot(sc_x.inverse_transform(X),sc_y.inverse_transform(y),color='red')
plt.plot(sc_x.inverse_transform(X),sc_y.inverse_transform(regressor.predict(X).reshape(-1, 1)),color='blue') # Reshape the predictions to a 2D array
plt.title('svr')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```
