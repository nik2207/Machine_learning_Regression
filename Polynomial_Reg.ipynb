#Importing Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#Reading a Csv File

```python
data_set = pd.read_csv("/content/RBC.csv")
```

```python
data_set.describe()
```

```python
x=data_set.iloc[:,0:-1].values
y=data_set.iloc[:,-1].values
```

#Training a Linear Regression on whole dataset

```python
from sklearn.linear_model import LinearRegression
Linear_reg =LinearRegression()
Linear_reg.fit(x,y)
```

#Training a Polynomial Regression on whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
Linear_reg_2 = LinearRegression()
Linear_reg_2.fit(x_poly,y)
```

#Visualize the Linear Regression Result..

```python
plt.scatter(x,y,color="Red")
plt.plot(x,Linear_reg.predict(x),color="Blue")
plt.title("RBC - Linear Regression")
plt.xlabel("Age")
plt.ylabel("Rbc")
plt.show
```

#Visualize the Polynomial Regression Result..

```python
plt.scatter(x,y,color="Red")
plt.plot(x,Linear_reg_2.predict(x_poly),color="Blue")
plt.title("RBC - Ploynomial Regression")
plt.xlabel("Age")
plt.ylabel("Rbc")
plt.show
```
